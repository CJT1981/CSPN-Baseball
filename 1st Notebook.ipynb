```python
# Sets
# Similar to lists and tuples, one big difference is that they do not carry
# duplicate values and values can not be accessed through indexes because 
# sets are unordered, they can be accessed through loops
daily_pints = [1,2,3]
print(type(daily_pints))
print(daily_pints)
```

```python
# example of sets only showing unique values even if you assign it with duplicates
num_set = {1,2,3,4,5,6,5,3,1} 
num_set_two = {4,5,6,7,8,9,6,5,7}
print(num_set)
print(num_set_two)
```

```python
# functions you can perform on sets
print(num_set | num_set_two) # this takes the combined unique values between the two sets
print(num_set & num_set_two) # this shows the matching values of both sets 
print(num_set - num_set_two) # this eliminates the shared values between both sets and 
                             # displays what's left of the first set
print(num_set ^ num_set_two) # this is going to show the values that are unique between both sets altogether 
```

# BMI Calculator

```python
# BMI = WEIGHT (LBS) * 703 / HEIGHT *HEIGHT (IN)
```

```python
name = input("Enter your name: ")
weight = int(input("Enter your weight in pounds: "))

height = int(input("Enter your height in inches: "))

bmi = (weight * 703) / (height * height)

print('Your BMI is: ', bmi)
if bmi > 0:
    if(bmi > 18.5):
        print(name + ", you are underweight.")
    elif(bmi <= 24.5):
        print(name + ", you are normal weight.")
    elif(bmi < 29.9):
        print(name + ", you are overweight.")
    elif(bmi < 34.9):
        print(name + ", you are obese.")
    elif(bmi < 39.9):
        print(name + ", you are severly obese.")
    else:
        print(name + ", you are morbidly obese.")
else:
    print("Enter valid input")
```

# Scraping Data from a Real Website + Pandas

```python
from bs4 import BeautifulSoup
import requests
url = "https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue"

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/125.0 Safari/537.36"
}

page = requests.get(url, headers=headers)

soup = BeautifulSoup(page.text, "html")

print(soup.prettify())
```

```python
print(soup)
```

```python
# We want to pull info from the tables so we are searching through all the tables to find the one we want
soup.find_all('table')
```

```python
soup.find_all('table')[1]
```

```python
#<table class="wikitable sortable jquery-tablesorter">
#</table>
# in the above cell, we used 1 to specify we wanted the first table, here is another way to pull it using the
# class name
soup.find_all('table', class_ = 'wikitable sortable')
```

```python
# We make a variable to hold the data of just the first table
table = soup.find_all('table')[0]
print(table)
```

```python
# first step to get the category names, the names are under the tag 'th', i.e. we used find all 'th'
world_titles = table.find_all('th')
print(world_titles)
```

```python
# now we are going to remove all the extra stuff around the names bc thats what we want
world_table_titles = [title.text.strip() for title in world_titles]
print(world_table_titles)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = world_table_titles)
df
```

```python
# there are two tags, one is the 'tr' that contains the entire row, but 'td' will contain the individual elements 
# of information we want and that is why we will use td and tr
column_data = table.find_all('tr')
```

```python
# in the above cell we are using getting all the rows of data
# in this cell we are going through the rows and finding all the data and appending it to the dataframe
for row in column_data[1:]: # <-- [1:] was necessary because the 1st row was empty and throwing errors
    row_data = row.find_all('td') # <-- getting all the data throuhg td tags
    individual_row_data = [data.text.strip() for data in row_data] # <-- cleaning up the data
    # print(individual_row_data)
    length = len(df)
    df.loc[length] = individual_row_data
```
